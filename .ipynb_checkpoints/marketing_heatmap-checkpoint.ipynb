1. Word2Vec을 이용하여 특정 속성과 코사인 거리가 짧은 단어 리스트 top 50
- ex) 속성1과 관련된 토큰이 50개, 속성2와 관련된 토큰이 50개 ....
<br>
<br>
2. 속성1부터 속성12까지, 총 50*12개의 토큰의 합집합을 구한 뒤, P(토큰)을 계산함
<br>
<br>
3. 각 속성에 대해 P(속성|토큰)을 계산함 (코사인 거리의 역수를 normalize함)
<br>
<br>
4. 특정 리뷰 문장이 있을 때, 토큰화를 실시하고, 토큰의 합집합에 속한 토큰만 남겨놓음
<br>
<br>
5. 각 토큰에 대해 P(속성|토큰) = P(토큰|속성)*P(속성)/P(토큰) 을 계산 —> 정규가 아무 속성에도 속하지 않는 경우, normalize 하면 확률값이 인위적으로 커질 수 있음을 우려했는데, 생각해보니 3번에서 토큰의 합집합만 고려하면 괜찮을 것 같음
<br>
<br>
6. 리뷰 문장에서 추출된 토큰에 대해 각각 P(속성|토큰)의 확률을 구하고, 이를 리뷰에 있는 모든 토큰에 대해 더한 뒤 다시 normalize 해서 P(속성|리뷰)를 계산
<br>
<br>
7. 몇 개의 리뷰를 예제로 보여줌

### 1) Import Packages

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from scipy import spatial
import numpy as np
import pandas as pd
import seaborn as sns
from collections import Counter
```

### 2) 사전에 학습된 word2vec 모델 불러오기 ( = `loaded_model` )

```python
loaded_model = KeyedVectors.load_word2vec_format("aspect_dict_w2v") # 모델 로드
```

### 3) 속성 리스트 11개
- 자주 등장하지 않는 속성은 제외시켜서, 최종 11개
- "속성명" vs "word2vec검색용 이름"
    - 속성명 : ['가격','광택','구성','디자인','맛','성분','용기','트렌드','표현','품질','휴대']
    - word2vec검색용 이름 : ['싸다','물광','알차다','디자인','맛','순하다','양','트렌드','표현','품질','휴대']

```python
aspect_list = ['싸다','물광','알차다','디자인','맛','순하다','양','트렌드','표현','품질','휴대']
```

### 4) 속성 벡터 : `aspect_vector`

```python
aspect_vector = np.array([loaded_model[i] for i in aspect_list])
```

```python
print('word2vec embedding vector의 차원 : ',aspect_vector[0].shape)
print('가격(검색명=싸다) 의 속성 벡터 : \n',aspect_vector[0])
```

### 5) `P_t_given_a_func` 함수

$P(\text{타겟}|\text{속성})$ = $\frac{\exp(d(\text{타겟},속성_k))}{\sum_{i=1}^{12}\exp(d(\text{타겟},속성_i))}$

- input : "token 단어" & word2vec 모델
- output : "token 단어"와 "11가지 속성"사이의 **코사인 유사도**

```python
def P_t_given_a_func(token,model):
    cossim_list=[]
    for av in aspect_vector:
        cossim=1 - spatial.distance.cosine(av, model[token])
        cossim_list.append(np.exp(cossim))
    cossim_list/=sum(cossim_list)
    return cossim_list
```

```python
P_t_given_a_func('촉촉하다',loaded_model)
```

### 6) 속성별로 가장 유사한 50개의 단어

```python
aspect_top50={}
aspect_top50_total=[]

for aspect in aspect_list:
    # Top 50인데 70으로하는 이유?
    # --> '한 글자'짜리 제외하기 위해!
    # --> 넉넉하게 Top 70까지 구한 뒤, '한 글자' 단어를 제외한 것 중 Top 50구하기
    top50_word_sim=loaded_model.most_similar(aspect,topn=70)
    top50_word=[x[0] for x in top50_word_sim]
    top50_word=[x for x in top50_word if len(x)>1]
    top50_word=top50_word[:50]
    aspect_top50[aspect]=top50_word
    aspect_top50_total.extend(top50_word)
```

```python
print('가격(검색명=싸다)와 가까운 Top 50 단어 : \n',aspect_top50['싸다'])
```

- [합집합] 11개 속성 & Top 50 = 550개의 단어
- BUT, 겹치는게 있어서 **550 $\rightarrow$ 513개**

```python
print("합집합 with 중복 :",len(aspect_top50_total))
print("합집합 without 중복 :",len(set(aspect_top50_total)))
```

- 합집합 list : `aspect_top50_total` ( 513개 )

```python
aspect_top50_total=list(set(aspect_top50_total))
```

### 7) $ P(\text{토큰})$ 계산 

```python
from ast import literal_eval
from itertools import chain
review_df = pd.read_csv('df_for_cbs.csv',usecols=["review_split", "nva"])

review_df['review_split']=review_df['review_split'].apply(literal_eval)
review_df['nva']=review_df['nva'].apply(literal_eval)
```

```python
total_words = list(chain(*review_df['nva']))
total_word_count = Counter(total_words)
```

```python
token_count=dict()
P_token=dict()

for token in aspect_top50_total:
    token_count[token]=total_word_count[token]
    
total_count = sum(token_count.values())

for token in aspect_top50_total:
    P_token[token]=token_count[token]/total_count
```

### 9) $P(\text{속성}|\text{타켓})=\frac{P(\text{타겟}|\text{속성}) \cdot P(\text{속성})}{P(\text{타켓})}$ 
- a) $P(\text{타겟}|\text{속성})$
- b) $P(\text{속성})$
- c) $P(\text{타켓})$

a) $P(\text{타겟}|\text{속성})$

```python
P_t_given_a_func('예민',loaded_model)
```

b) $P(\text{속성})$

```python
P_aspect=1/len(aspect_list)
```

c) $P(\text{타켓})$

```python
P_token['예민']
```

$\frac{P(\text{타겟}|\text{속성}) \cdot P(\text{속성})}{P(\text{타켓})}$

```python
token_posterior=dict()
token_posterior_normalized=dict()

for token in aspect_top50_total:
    token_posterior[token] = (P_t_given_a_func(token,loaded_model)*P_aspect)/P_token[token]
    token_posterior_normalized[token]=token_posterior[token]/sum(token_posterior[token])
```

### 11) Example 

`NVA` : Noun Verb Adjective를 뽑아내는 함수

- input : review 리스트들
    - ex) ['리뷰1','리뷰2' ... ]
- output : [[리뷰1의 N,V,A들], [리뷰2의 N,V,A들] ... ]

```python
import konlpy

def NVA(x):    
    total_words=[]
    for i, document in enumerate(list(set(x))):
        okt = konlpy.tag.Okt()
        clean_words = []
        for word in okt.pos(document, stem=True): 
            if word[1] in ['Noun', 'Verb', 'Adjective']: 
                clean_words.append(word[0])
        total_words.append(clean_words)
    result = []
    for x in total_words:
        result.extend(x)
    return result
```

```python
def review_heatmap(review,normalize=True):
    # 1) Review -> NVA 의 Reveiw -> NVA + Top50에 속하는 Review
    nva=NVA([review])
    nva_filtered=[word for word in nva if word in aspect_top50_total]
    print('[ 리뷰 ] : \n',review,'\n')
    print('[ 리뷰 (형/명/부) ]: \n',nva,'\n')
    print('[ 리뷰 (속성 단어 Top50) ]: \n',nva_filtered,'\n')
    
    if len(nva_filtered)==0:
        return 'NO TOKEN FOUND'
    
    # 2) Review의 posterior 구하기 (normalize = True or False)
    nva_filtered_posterior = [token_posterior_normalized[word] for word in nva_filtered]
    posterior_heatmap_value = np.sum(nva_filtered_posterior,axis=0)
    if normalize:
        posterior_heatmap_value=posterior_heatmap_value/sum(posterior_heatmap_value)
    
    # 3) 시각화를 위한 dataframe 만들기
    df_index=['가격','광택','구성','디자인','맛','성분','용기','트렌드','표현','품질','휴대']
    viz_df = pd.DataFrame(pd.Series(posterior_heatmap_value,index=df_index),columns=['value'])
    
    # 4) heatmap bar의 최소/최대값
    v_max_val=np.max(viz_df['value'])*1.2
    v_min_val=np.min(viz_df['value'])*0.8
    
    # 5) heatmap
    sns.heatmap(viz_df, annot=True, linewidths=.5,cmap='RdBu_r',vmin=v_min_val, vmax=v_max_val)
```

```python
example_review = '한율 마스크는 저렴한 가격에 좋은 것 같습니다. 트러블을 진정시키는데에 좋은것 같아요'


review_heatmap(example_review)
```
